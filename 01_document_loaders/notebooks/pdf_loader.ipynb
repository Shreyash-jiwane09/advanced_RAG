#### Langchain pdf loader : https://reference.langchain.com/python/langchain-community/document_loaders/pdf/PyPDFLoader

```python
#import dependencies
from langchain_community.document_loaders import PyPDFLoader
from pathlib import Path
from pprint import pp
```

```python
# create the path for pdf file

file_path = Path("../knowledge-source/attention_is_all_you_need.pdf")

#convert file path to str
file_path_str = file_path.as_posix()

# Check whether the file exists or not
#file_path.exists() 
```

```python
# create a loader

pypdf_loader = PyPDFLoader(file_path=file_path_str,
                           mode='page')
```

```python
# load the documents

documents = pypdf_loader.load()
print(documents[0].page_content)

#number of documents
len(documents)
```

```python
pp(documents[0].metadata)
```

```python
# fetch the first page text content

print(documents[1].page_content)
```

```python
# metadata

pp(documents[4].metadata)
```

#### TesseractBlobParser : Parse for extracting text from images using the Tesseract OCR library.
#### RapidOCRBlobParser : Parser for extracting text from images using the RapidOCR library.

```python
from langchain_community.document_loaders.parsers import TesseractBlobParser, RapidOCRBlobParser
```

```python
# create pypdf instance which can extract images

pypdf_image_loader = PyPDFLoader(file_path=file_path.as_posix(),
                                 mode="page",
                                 extract_images=True,
                                 images_parser=RapidOCRBlobParser(),
                                 images_inner_format="html-img")
```

```python
# load documents with images

documents_with_images = pypdf_image_loader.load()
```

```python
page_with_image = documents_with_images[2]
page_with_image
```

```python
# length of the document

len(documents_with_images)
```

```python
print(page_with_image.page_content[-700:])
```

#### PDFMinerLoader
    Load and parse a PDF file using 'pdfminer.six' library.

    This class provides methods to load and parse PDF documents, supporting various configurations such as handling password-protected files, extracting images, and defining extraction mode. It integrates the pdfminer.six library for PDF processing and offers both synchronous and asynchronous document loading.

```python
from langchain_community.document_loaders import PDFMinerLoader
```

```python
# create the miner loader

pdfminer_loader = PDFMinerLoader(file_path=file_path.as_posix(),
                                 mode="page",
                                 extract_images=True,
                                 images_parser=RapidOCRBlobParser(),
                                 images_inner_format="html-img")
```

```python
# load the documents

documents_with_images = pdfminer_loader.load()
```

```python
# image text in pdf

print(documents_with_images[2].page_content[-450:])
```

```python
# text from table

print(documents_with_images[5].page_content)
```

```python
pp(documents_with_images[0].metadata)
```

```python
for doc in documents_with_images:
    if doc.metadata["page"] == 7:
        print(doc.page_content)
        break
```
